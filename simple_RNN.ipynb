```python
import torch
import torch.nn as nn
```

```python
rnn = nn.RNN(
    input_size=10,   # feature size
    hidden_size=20,  # memory size
    num_layers=1,
    batch_first=True
)

x = torch.randn(32, 5, 10)  # batch, sequence length, features
output, hidden = rnn(x)

print(output.shape)  # (32, 5, 20)
```
